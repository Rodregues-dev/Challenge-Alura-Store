### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# prompt: Nesta primeira análise, você deve calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento.

faturamento_total_loja1 = loja['Preço'].sum()
faturamento_total_loja2 = loja2['Preço'].sum()
faturamento_total_loja3 = loja3['Preço'].sum()
faturamento_total_loja4 = loja4['Preço'].sum()

print(f"Faturamento total da Loja 1: R${faturamento_total_loja1:,.2f}")
print(f"Faturamento total da Loja 2: R${faturamento_total_loja2:,.2f}")
print(f"Faturamento total da Loja 3: R${faturamento_total_loja3:,.2f}")
print(f"Faturamento total da Loja 4: R${faturamento_total_loja4:,.2f}")
```

```python
# prompt: Faça um gráfico para ilustrar esses faturamentos

import matplotlib.pyplot as plt

faturamentos = [faturamento_total_loja1, faturamento_total_loja2, faturamento_total_loja3, faturamento_total_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.ylabel('Faturamento (R$)')
plt.title('Faturamento Total por Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Vendas por Categoria


```python
# prompt: Neste passo, você deve calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

# Agrupar por categoria e contar o número de vendas em cada loja
vendas_por_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("Vendas por Categoria na Loja 1:")
print(vendas_por_categoria_loja1)
print("\nVendas por Categoria na Loja 2:")
print(vendas_por_categoria_loja2)
print("\nVendas por Categoria na Loja 3:")
print(vendas_por_categoria_loja3)
print("\nVendas por Categoria na Loja 4:")
print(vendas_por_categoria_loja4)

# Opcional: Visualizar as vendas por categoria para cada loja
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

vendas_por_categoria_loja1.plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Vendas por Categoria - Loja 1')
axes[0, 0].set_xlabel('Categoria')
axes[0, 0].set_ylabel('Número de Vendas')
axes[0, 0].tick_params(axis='x', rotation=45)

vendas_por_categoria_loja2.plot(kind='bar', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Vendas por Categoria - Loja 2')
axes[0, 1].set_xlabel('Categoria')
axes[0, 1].set_ylabel('Número de Vendas')
axes[0, 1].tick_params(axis='x', rotation=45)

vendas_por_categoria_loja3.plot(kind='bar', ax=axes[1, 0], color='salmon')
axes[1, 0].set_title('Vendas por Categoria - Loja 3')
axes[1, 0].set_xlabel('Categoria')
axes[1, 0].set_ylabel('Número de Vendas')
axes[1, 0].tick_params(axis='x', rotation=45)

vendas_por_categoria_loja4.plot(kind='bar', ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Vendas por Categoria - Loja 4')
axes[1, 1].set_xlabel('Categoria')
axes[1, 1].set_ylabel('Número de Vendas')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# prompt: Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

# Calcular a média das avaliações para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação da Loja 4: {media_avaliacao_loja4:.2f}")

# Opcional: Visualizar a média de avaliação por loja
avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 6))
plt.bar(lojas, avaliacoes, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação por Loja')
plt.ylim(0, 5) # Definir o limite do eixo y para avaliações (geralmente de 1 a 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# prompt: Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

# Concatenar os dataframes para facilitar a análise
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

df_todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Agrupar por loja e produto para contar as vendas
vendas_por_loja_produto = df_todas_lojas.groupby(['Loja', 'Produto'])['Produto'].count().reset_index(name='Número de Vendas')

# Identificar o produto mais vendido e menos vendido em cada loja
produtos_destaque_por_loja = pd.DataFrame()

for loja_nome in vendas_por_loja_produto['Loja'].unique():
    df_loja = vendas_por_loja_produto[vendas_por_loja_produto['Loja'] == loja_nome]

    # Produto mais vendido
    produto_mais_vendido = df_loja.loc[df_loja['Número de Vendas'].idxmax()]

    # Produto menos vendido (considerando apenas produtos que foram vendidos)
    if not df_loja.empty:
        produto_menos_vendido = df_loja.loc[df_loja['Número de Vendas'].idxmin()]
    else:
        produto_menos_vendido = {'Loja': loja_nome, 'Produto': 'Nenhum', 'Número de Vendas': 0}


    produtos_destaque_por_loja = pd.concat([produtos_destaque_por_loja,
                                            pd.DataFrame([produto_mais_vendido]),
                                            pd.DataFrame([produto_menos_vendido])],
                                           ignore_index=True)

produtos_destaque_por_loja['Tipo'] = ['Mais Vendido', 'Menos Vendido'] * 4

print("Produtos mais e menos vendidos por Loja:")
print(produtos_destaque_por_loja)

# Visualização
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for i, loja_nome in enumerate(vendas_por_loja_produto['Loja'].unique()):
    df_loja = vendas_por_loja_produto[vendas_por_loja_produto['Loja'] == loja_nome]

    # Ordenar para melhor visualização
    df_loja_sorted = df_loja.sort_values(by='Número de Vendas', ascending=False)

    # Selecionar os top 5 mais vendidos e os top 5 menos vendidos (se houver mais de 10 produtos únicos)
    if len(df_loja_sorted) > 10:
      df_plot = pd.concat([df_loja_sorted.head(5), df_loja_sorted.tail(5)])
    else:
      df_plot = df_loja_sorted

    axes[i].bar(df_plot['Produto'], df_plot['Número de Vendas'], color=['skyblue' if vendas > df_plot['Número de Vendas'].mean() else 'salmon' for vendas in df_plot['Número de Vendas']])
    axes[i].set_title(f'Vendas por Produto - {loja_nome}')
    axes[i].set_xlabel('Produto')
    axes[i].set_ylabel('Número de Vendas')
    axes[i].tick_params(axis='x', rotation=90)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
# prompt: Neste passo, vamos calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

# Calcular o custo médio de frete para cada loja
custo_medio_frete_loja1 = loja['Frete'].mean()
custo_medio_frete_loja2 = loja2['Frete'].mean()
custo_medio_frete_loja3 = loja3['Frete'].mean()
custo_medio_frete_loja4 = loja4['Frete'].mean()

print(f"Custo médio de frete da Loja 1: R${custo_medio_frete_loja1:,.2f}")
print(f"Custo médio de frete da Loja 2: R${custo_medio_frete_loja2:,.2f}")
print(f"Custo médio de frete da Loja 3: R${custo_medio_frete_loja3:,.2f}")
print(f"Custo médio de frete da Loja 4: R${custo_medio_frete_loja4:,.2f}")

# Visualizar o custo médio de frete por loja
custos_frete = [custo_medio_frete_loja1, custo_medio_frete_loja2, custo_medio_frete_loja3, custo_medio_frete_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 6))
plt.bar(lojas, custos_frete, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.ylabel('Custo Médio do Frete (R$)')
plt.title('Custo Médio do Frete por Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
